```python
#!/usr/bin/python

import os, sys
from bs4 import BeautifulSoup
import re


def getRating(review):
    rating='NA'
    divReview = review.find('div',{'class':re.compile('i-stars')})
    if(divReview):
        imgAttr = divReview.find('img')['alt']
        temp = imgAttr.split(' ')
        rating = float(imgAttr[0])
    return rating
def getReview(review):
    review='NA'
    
def getUserIdandName(review):
    userId,userName = 'NA'
    a=review.find('a',{'class':re.compile('user-display-name')})
    if(a):
        userName=a.text
        link =a['href']
        temp = link.split('userid=')
        userId = temp[1]
        #print(userName,userId)
    
    return userId,userName

    

# Open a file
path = "/Volumes/Khushali Harddisk/YelpDataSet"
dirs = os.listdir( path )
fw=open('/Volumes/Khushali Harddisk/reviews.txt','w') # output file
# This would print all the files and directories
for directory in dirs:
    review_file = os.listdir(path+"/"+directory)
    #print(review_file)
    for i in range(0,len(review_file)):
        arr = review_file[i]
        soup = BeautifulSoup(open(path+"/"+directory+"/"+arr), "html.parser")
        reviews=soup.findAll('div', {'class':re.compile('review--with-sidebar')}) # get all the review divs
        restTag =soup.find('h1',{'class':'biz-page-title'})

        if(restTag):
            restName=restTag.text
            for review in reviews:

                #critic = getCritic(review)
                if(review):
                    rating = getRating(review)
                    userId,userName = getUserIdandName(review)
                print(restName.strip()+','+userId+','+userName+','+str(rating))
                fw.write(restName.strip()+','+userId+','+userName+','+str(rating)+'\n') # write to file 


fw.close()    
        #print(arr)

    

```
